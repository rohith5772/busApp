```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
# Send a GET request to the URL
url = 'http://www.hyderabadbusroutes.in/bus-info/593'
val = url.split('/')
busid = val[4]
print(busid)
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
px = busid
temp =  ""
# Find the table element and extract the table headers and rows
table = soup.find('table')
#headers = ['bus_id'] + ['stop_id']+['Arr_Time'] +['Dep_Time']+['Freq']+['From'] +['To']+['stop_timings']+['station_order']+['station_order']
rows = []
testrows = []
for i, tr in enumerate(table.find_all('tr')[2:], start=1):
    testrows.append([px]+ [td.text.strip() for td in tr.find_all('td')])
print(len(testrows))
for j, tr in enumerate(table.find_all('tr')[3:], start=1):
    print(j)
    if(j<len(testrows)):
        rows.append([px] + [testrows[j][2]]+ [temp]+ [temp]+[temp]+[testrows[1][2]]+[testrows[-1][2]]+[j]+[td.text.strip() for td in tr.find_all('td')])


# Write the data to a CSV file
filename = 'bus_info.csv'
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    #writer.writerow(headers)
    for row in rows:
        writer.writerow(row)

print(f'Data written to {filename} successfully.')

table = soup.find('table')
#headers = ['bus_id'] + ['stop_id']+['Arr_Time'] +['Dep_Time']+['Freq']+['From'] +['To']+['stop_timings']+['station_order']+['station_order']
rows = []
testrows = []
for i, tr in enumerate(table.find_all('tr')[2:], start=1):
    testrows.append([px]+ [td.text.strip() for td in tr.find_all('td')])
print(len(testrows))
k=j
for j, tr in enumerate(table.find_all('tr')[3:], start=1):
    print(j)
    if(j<len(testrows)):
        rows.append([px] + [testrows[j][2]]+ [temp]+ [temp]+[temp]+[testrows[-1][2]]+[testrows[1][2]]+[k]+[td.text.strip() for td in tr.find_all('td')])
        k=k-1
# Write the data to a CSV file
rows.reverse()
filename = 'bus_info_return.csv'
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    #writer.writerow(headers)
    for row in rows:
        writer.writerow(row)

print(f'Data written to {filename} successfully.')
# open the first CSV file and read its contents
with open('bus_info.csv', 'r') as file1:
    reader = csv.reader(file1)
    data1 = list(reader)

# open the second CSV file and read its contents
with open('bus_info_return.csv', 'r') as file2:
    reader = csv.reader(file2)
    data2 = list(reader)

# merge the two lists of data
merged_data = data1 + data2

# write the merged data to a new CSV file
with open('merged.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(merged_data)

with open('merged.csv', 'r') as infile, open('output.csv', 'w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # iterate over each row in the input CSV file
    for row in reader:

        # write the first n-3 columns to the output CSV file
        writer.writerow(row[:-4])


with open('busApp.csv', 'r') as file1:
    reader = csv.reader(file1)
    data1 = list(reader)

# open the second CSV file and read its contents
with open('output.csv', 'r') as file2:
    reader = csv.reader(file2)
    data2 = list(reader)

# merge the two lists of data
merged_data = data1 + data2

# write the merged data to a new CSV file
with open('busApp.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(merged_data)
```
